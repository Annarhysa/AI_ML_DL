```python
import numpy as np
import math as m

def unitStep(x):
  if (x>=0.5):
    return 1
  else:
    return 0

def model(x, w, b):
  z = np.dot(w,x) + b
  func = 1/(1+m.exp(-z))
  y = unitStep(func)
  return y

def Logical_NOT(x):
  w = np.array([-1])
  x = np.array(x)
  b = 0
  y = model(x,w,b)
  return y

print(f"NOT(1) = {Logical_NOT(1)}")
print(f"NOT(0) = {Logical_NOT(0)}")
```

```python
def Logical_NAND(x):
  w = np.array([-2,-1])
  x = np.array(x)
  b = 2
  y = model(x,w,b)
  return y

print(f"NAND(1,0) = {Logical_NAND([1,0])}")
print(f"NAND(0,1) = {Logical_NAND([0,1])}")
print(f"NAND(1,1) = {Logical_NAND([1,1])}")
print(f"NAND(0,0) = {Logical_NAND([0,0])}")
```

```python
def Logical_NOR(x):
  w = np.array([-1,-1])
  x = np.array(x)
  b = 0
  y = model(x,w,b)
  return y

print(f"NOR(1,0) = {Logical_NOR([1,0])}")
print(f"NOR(0,1) = {Logical_NOR([0,1])}")
print(f"NOR(1,1) = {Logical_NOR([1,1])}")
print(f"NOR(0,0) = {Logical_NOR([0,0])}")
```
